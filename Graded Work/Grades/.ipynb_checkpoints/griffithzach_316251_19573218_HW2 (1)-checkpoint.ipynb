### [20 points] Problem 1

To play a game, you have a box containing 20 fair [four-sided dice](https://en.wikipedia.org/wiki/Four-sided_die), with faces `{1,2,3,4}`.  This bag also contains 15 fair six-sided dice (faces `{1,2,3,4,5,6}`) and 5 fair twenty-sided dice (faces `{1,2,3,4,...,19,20}`).  Call these 3 classes of die "Four", "Six" and "Twenty" (or F, S, and T, for short).  You grab one die at random from the box.  

Work the following problems by hand (and LaTeX your solutions) unless otherwise stated (but don't be afraid to simulate to check your result!).

**Part A**: You roll your die one time.  What is the probability of the event $R_6$, that you roll a 6?  Include at least 4 decimal places in your final answer.

**Solution**:

This is a conditional probability. 

$$P(F) =\frac{20}{40}=\frac{1}{2},\quad P(S)=\frac{15}{40}=\frac{3}{8},\quad P(T)=\frac{5}{40}=\frac{1}{8}$$
\begin{align*}
\\ P(R_6)  &=P(F\cap R_6) + P(S\cap R_6) + P(T \cap R_6) \\[1em]
& \quad \text{Since the four sided dice doesn't have a 6, probability is zero} \\[1em]
&= 0 + P(S)P(R_6|S) + P(T)P(R_6|T) \\[1em] 
& \quad \text{There are 15 of 'S' Type, so 15 sixes out of 90 total faces, for 'T', there are 5 sixes out of 100 total faces} \\[1em]
&= (\frac{3}{8})(\frac{1}{6}) + (\frac{1}{8})(\frac{1}{20}) = \frac{11}{160} \\[1em]
P(R_6) &= \boxed{0.06875}
\end{align*}

**Part B**: Suppose you roll a 6.  What is the probability that the die you chose from the box is a Six-sided die? 

**Solution**:
\begin{align*}
&\text{This can be solved with Bayes Theorem} \\[1em] P(S|R_6)&=\frac{P(S)P(R_6|S)}{P(R_6)} & \\[1em] &=\frac{\frac{3}{8}*\frac{1}{6}}{\frac{11}{160}} \\[1em] &= \frac{\frac{1}{16}}{\frac{11}{160}} = \frac{10}{11} = \boxed{90\%}
\end{align*}

**Part C:**  Are the events $R_6$ and $S$ independent?  Justify your answer **using math**.

**Solution:** 

$R_6$ and $S$ are ***NOT*** independent because: $$P(R_6|S) \neq P(R_6)$$

$P(R_6|S)$ is equal to $\frac{1}{6}$ and $P(R_6)$ is equal to $\frac{11}{160}$ These two values would need to be equal in order for the events to be independent.

**Part D:** Suppose you roll your die $n$ times and record each outcome.  Obviously, if you roll 5 or 6 you know you did *not* pick a Four-sided die, and if you roll a 7 or higher, you must have picked a Twenty-sided die.

But, if you roll $n$ times in a row and get only 1, 2, 3 or 4, you would have a stronger belief that you in fact grabbed one of the Four-sided dice.  Let $E_n$ be the event that you roll $n$ rolls of 1, 2, 3 or 4 in a row.  For example, one outcome in $E_3$ is the sequence 2, 3, 1.

Given that you roll a single value of 1, 2, 3 or 4, what is the probability that you have grabbed a Four-sided die?  That is, calculate **by hand** $P(F \mid E_1)$. 

**Solution:**
\begin{align*}
P(F|E_1) &= \frac{P(F \cap E_1)}{P(E_1)} = \frac{P(F)P(E_1|F)}{P(E_1)} \\[1em]
\text{Finding } P(E_1): P(E_1) &= P(F \cap E_1) + P(S \cap E_1) + P(T \cap E_1) \\[1em]
&= P(F)P(E_1|F) + P(S)P(E_1|S) + P(T)P(E_1|T) \\[1em]
&= (\frac{1}{2} * 1) + (\frac{15}{40} * \frac{4}{6}) + (\frac{5}{40} * \frac{4}{20}) \\[1em]
&= \frac{1}{2} + \frac{1}{4} + \frac{4}{160} \\[1em]
P(E_1)&= \frac{31}{40} \\[1em]
&\text{Now I plug into the top equation:}\\[1em]
P(F|E_1)&=\frac{\frac{1}{2}}{\frac{31}{40}}\\[1em]
&=\frac{40}{62} = \boxed{64.51\%}
\end{align*}

**Part E**:  First, work out **by hand** a formula for the probability in general that you are rolling a Four-sided die, given that you have rolled $n$ values in a row that are 1, 2, 3 or 4 (that is, $E_n$).  So you want a formula for $P(F \mid E_n)$ in terms of $n$.

Then, define and code a function `prob_four` to calculate $P(F \mid E_n)$.  This function should only take one positive integer argument, $n$.

Verify that your function is performing as expected by checking your answer to **Part D** when $n=1$.

**Solution**:
\begin{align*}
P(F|E_n) &= \frac{P(F \cap E_n)}{P(E_n)} = \frac{P(F)P(E_n|F)}{P(E_n)} \\[1em]
\text{From above}: P(E_1) &= P(F \cap E_1) + P(S \cap E_1) + P(T \cap E_1) \\[1em]
&= P(F)P(E_1|F) + P(S)P(E_1|S) + P(T)P(E_1|T) \\[1em]
&=\boxed{ (\frac{1}{2} * 1) + (\frac{15}{40} * \frac{4}{6}) + (\frac{5}{40} * \frac{4}{20})} \\[1em]
&\text{Then I found } P(E_2)\\[1em]
P(E_2) &= P(F \cap E_2) + P(S \cap E_2) + P(T \cap E_2) \\[1em]
&= P(F)P(E_2|F) + P(S)P(E_2|S) + P(T)P(E_2|T) \\[1em]
&=\boxed{ (\frac{1}{2} * 1 * 1) + \frac{15}{40}(\frac{4}{6} * \frac{4}{6}) + \frac{5}{40}( \frac{4}{20}*\frac{4}{20})} \\[1em]
&\text{From here I saw the pattern in the boxed sections, Then simplified } P(E_n) \\[1em]
P(E_n) &= \frac{3}{8}(\frac{4}{6})^n + \frac{1}{8}(\frac{4}{20})^n + \frac{1}{2} \\[1em]
&\text{plugging in to top equation;}\\[1em]
P(F|E_n) &= \frac{\frac{1}{2}(1)}{ \frac{3}{8}(\frac{4}{6})^n + \frac{1}{8}(\frac{4}{20})^n + \frac{1}{2}} = \frac{1}{2[\frac{3}{8}(\frac{4}{6})^n + \frac{1}{8}(\frac{4}{20})^n + \frac{1}{2}]} \\[1em]
&= \boxed{\frac{1}{\frac{3}{4}(\frac{4}{6})^n + \frac{1}{4}(\frac{4}{20})^n + 1}}
\end{align*}

```python
def prob_four(n):
    denom = 0
    for i in range(1,n+1):
        denom = (3/4*(4/6)**i) + (1/4*(4/20)**i) + 1
        print("P(F|E_n), n =",i,": ",(1/denom)*100,"%")

prob_four(2)
```

```python
def miniopoly_turn(state, cash, bonus=500):
    # roll two dice
    a = np.random.randint(1, 7)
    b = np.random.randint(1, 7)
    roll = a + b
    
    # Handle special positions
    doNothing = [1,7,10]
    drawTrigger = [2,8,13,17,25,33]
    other = [0,20,30]
    special = doNothing + drawTrigger + other
    
    # Handle paid cells, decide amount to pay/paid
    payOrGetPaid = [-75, -50, 50, 100] 
    payBank = [i for i in range(40) if i not in special] 
    cardResult = np.random.choice(payOrGetPaid) 
    
    if(state == 30): # If player is already in Jail
        if(a == b): # If roll is doubles
            state = 10 + roll
        else: # Does not roll doubles (return still in jail)
            cash -= 25
            if(cash <= 0): # If jail causes bankruptcy
                cash = 0
            return state, cash
    else: # Continue turn, handle passed go
        state += roll
        if (state > 39): 
            cash += 200
            state -= 40 
    
    if(state == 30): # Lands in jail 
        return state, cash
    elif (state in drawTrigger): # Lands in draw cell
        cash+=cardResult
    elif (state == 20): # Lands in bonus cell
        cash+=bonus
    elif (state in doNothing): # Lands in a do nothing cell
        return state, cash
    elif (state in payBank): # Lands in any other place.
        cash -= state
        
    if(cash <= 0): # Out of cash after turn
        cash = 0
        
    return state, cash

output = miniopoly_turn(0, 200)
print("Cell Position:", output[0])
print("Cash: $",output[1])
```

Make a **density** histogram of the player's cash amount at the end of the game under the both the 20-turn and 40-turn scenarios.
```python
#-----------------------------------------------------------------------------------------------
#  I am using this function to simulate instead of simulating over and over with different loops.
#  loops, this function includes all of the functionality for Parts: B, C and D
#
#  Function takes inputs: 
#     desired number of samples, 
#     specific number of miniopoly turns, 
#     part: either B, C or D. i.e: Problem 4, Part: (part)
#
#  Depending on the part of Problem 4, the function will change it's functionality
#-----------------------------------------------------------------------------------------------
def simulate_miniopoly(num_samples, turns, part):
    
    # Counters and inputs to miniopoly function
    games, turnCount = 0, 0
    state, cash = 0, 200
    
    # Store ending cash ammounts For parts B or C
    gameData = []
    
    # For part C
    bankruptCount = 0
    
    # For part D
    # Counting for probability calculation
    jailCount = 0
    wonLottCount = 0
    jail2x = 0
    noLottery = 0
    # to calculate 𝑃(Jail at least twice ∩ no lottery wins) 
    j2xAndNL = 0
    # to calculate 𝑃(Bankrupt ∩ j2xAndNL) 
    BRandConditions = 0
    
    while(games < num_samples): # num_samples = input(number of games)
        turn = miniopoly_turn(state, cash)
        turnCount+=1
        
        # For part D
        if(turn[0]==30):
            jailCount += 1
        if(turn[0]==20):
            wonLottCount += 1
        
        if(turn[1] == 0 or turnCount == turns): # End of a game (turns as fn input)
            games+=1
            turnCount = 0

            if(part == 'B'): # Store ending cash ammounts
                gameData.append(turn[1])
                
            if(part == 'C'):
                if(turn[1] == 0):
                    bankruptCount+=1 # Count bankruptcies
                    
            if(part == 'D'):
                # Count num times player had: (Jail at least twice ∩ no lottery wins)
                if(jailCount >= 2 and wonLottCount == 0):
                    j2xAndNL += 1
                    # Count num times (Bankrupt ∩ j2xAndNL)
                    if(turn[1]==0):
                        BRandConditions += 1
                
            state, cash = 0, 200
            # For part D
            jailCount, wonLottCount = 0, 0 # Reset counts after game
            continue
            
        state, cash = turn[0], turn[1]
    
    # Outputs corresponding to part
    if(part == 'B'):
        return gameData
    
    if(part == 'C'):
        return bankruptCount
    
    if(part == 'D'):
        # Calculating probabilites of numerator
        # and denominator for conditional probability calculation
        # 𝑃(𝐵𝑅 ∩ 𝐶𝑜𝑛𝑑𝑖𝑡𝑖𝑜𝑛𝑠) / 𝑃(𝐶𝑜𝑛𝑑𝑖𝑡𝑖𝑜𝑛𝑠)
        P_BRandConditions = (BRandConditions/10000)
        P_j2xAndNL = (j2xAndNL/10000)

        # Probability of 𝑃(𝐵𝑅 | 𝐶𝑜𝑛𝑑𝑖𝑡𝑖𝑜𝑛𝑠)
        desiredProb = (P_BRandConditions / P_j2xAndNL) * 100
        return desiredProb
```

Suppose you roll two fair six-sided dice.  Let $C$ be the event that the two rolls are _close_ to one another in value, in the sense that they're either equal or differ by only 1. 

**Part A**:  Compute $P(C)$ by hand. 


**Solution:**
Out of 36 possible outcomes of rolling two dice, There are 16 outcomes in which the two dice had a face value within 1 of eachother. $$ P(C) = \frac{16}{36} = 0.4\overline{4} = \boxed{44.4\%} $$

**Part C**: In the in-class notebook associated with Lecture 4 (nb04) we plotted a running estimate of the probability of an event as a function of the number of trials in our simulation.  Write code to run at least 10 independent simulations of at least 50,000 trials each to estimate $P(C)$ and plot their running estimate curves on the same set of axes. **Hint**: This is a lot of computation, so try to leverage Numpy as much as possible so that your code doesn't run forever. 

```python
die = np.array([1,2,3,4,5,6])
num_rolls = 100000
fig, ax = plt.subplots(figsize=(40,30))

sims = 0
while(sims < 10):
    rolls1 = np.random.choice(die, size=num_rolls)
    rolls2 = np.random.choice(die, size=num_rolls)
    
    dif = abs(rolls1-rolls2) # List of absolute value of differences
    compare = dif <= 1 # Bool list of close rolls
    
    probs = np.cumsum(compare)/np.arange(1,len(compare)+1)
    ax.plot(probs)
    sims+=1
    
ax.set_title("Running Estimate of P(C)", fontsize=40)
ax.set_xlabel("Number of Rolls", fontsize=30)
ax.set_ylabel("Estimate of Probability", fontsize=30)
ax.yaxis.set_ticks(np.arange(0,1,0.05))
ax.tick_params(axis = 'y', labelsize = 22)
ax.tick_params(axis = 'x', labelsize = 22)
ax.grid(True, alpha=0.25)
```

**Part D**: Comment on the behavior of the running estimates as the number of trials increases. 

**Solution:**

Within the first 10,000 rolls, the variance between each of the 10 running estimates is fairly large, but as the number of trials increases, all 10 curves converge to 0.44 which is the theoretical answer to $P(C)$
